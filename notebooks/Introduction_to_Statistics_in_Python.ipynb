####**5. Types de statistiques**
+ Il existe 2 branches principales de la statistique: 
  + **la statistique descriptive et la statistique inférentielle**. 
  
+ **Les statistiques descriptives se concentrent sur la description et la synthèse des données disponibles**. 

+ Après avoir demandé à quatre amis comment ils se rendent au travail, nous pouvons voir que 50% d’entre eux se rendent au travail en voiture, 25% en bus et 25% à vélo. 

+ Ce sont des exemples de **statistiques descriptives**. 

+ **Les statistiques inférentielles utilisent les données disponibles, appelées données d’échantillonnage, pour faire des inférences au sujet d’une population plus importante**. 

+ Nous pourrions utiliser des statistiques inférentielles pour déterminer quel pourcentage de personnes conduisent pour se rendre au travail en fonction de nos données d’échantillonnage.

####**6. Types de données**
+ Il existe deux principaux types de données. 

+ **Les données numériques ou quantitatives sont constituées de valeurs numériques**. 

+ **Les données catégorielles ou qualitatives sont constituées de valeurs appartenant à des groupes distincts**. 

+ Il est important de noter que ce ne sont pas les deux seuls types de données qui existent - il y en a d’autres aussi, mais nous allons nous concentrer sur ces deux-là. 

+ **Les données numériques** peuvent être séparées en **données continues et discrètes**. 

+ Les données numériques continues sont souvent des quantités qui peuvent être mesurées, comme la vitesse ou le temps. 

+ Les données numériques discrètes sont généralement des données de comptage, comme le nombre d’animaux de compagnie ou le nombre de colis expédiés. 

+ **Les données catégorielles** peuvent être **nominales ou ordinales**. 

+ Les données catégorielles nominales sont constituées de catégories sans ordre inhérent, comme le statut matrimonial ou le pays de résidence. 

+ Les données catégorielles ordinales ont un ordre inhérent, comme une question d’enquête où vous devez indiquer le degré auquel vous êtes d’accord avec un énoncé.

####**7. Les données catégorielles peuvent être représentées sous forme de nombres**
+ Parfois, les variables catégorielles sont représentées à l’aide de nombres. 

+ Les personnes mariées et non mariées peuvent être représentées à l’aide de 1 et 0, ou une échelle d’accord peut être représentée par les numéros 1 à 5. 

+ Cependant, il est important de noter que cela n’en fait pas nécessairement des variables numériques.

####**8. Pourquoi le type de données est-il important?**
+ Il est important de pouvoir identifier les types de données, car le type de données avec lequel vous travaillez dictera les types de statistiques récapitulatives et de visualisations qui ont du sens pour vos données, c’est donc une compétence importante à maîtriser. 

+ Pour les données numériques, nous pouvons utiliser des statistiques sommaires comme la moyenne, et des graphiques comme les nuages de points, mais ceux-ci n’ont pas beaucoup de sens pour les données catégorielles.

####**9. Pourquoi le type de données est-il important?**
+ De même, des éléments tels que les comptages et les diagrammes à barres n’ont pas beaucoup de sens pour les données numériques.

####**4. Combien de temps les mammifères de cet ensemble de données dorment-ils généralement?**

+ Une façon de résumer les données est de répondre à la question suivante: 
  + Combien de temps les mammifères de cet ensemble de données dorment-ils généralement? 
  
+ Pour répondre à cette question, nous devons 
déterminer quelle est la valeur « typique » ou « centrale » des données. 
  
+ Nous discuterons de trois définitions différentes, ou mesures, du centre: moyenne, médiane et mode.

####**5. Mesures du centre: moyenne**
+ **La moyenne, souvent appelée moyenne, est l’une des façons les plus courantes de résumer les données.** 

+ Pour calculer la moyenne, nous additionnons tous les nombres d’intérêt et divisons par le nombre total de points de données, qui est de 83 ici. Cela nous donne 10 points 43 heures de sommeil. 

+ En Python, nous pouvons utiliser la fonction **np.mean**, en lui passant la variable d’intérêt.

####**6. Mesures du centre : médiane**

+ Une autre mesure du centre est la médiane. 

+ ***La médiane est la valeur où 50 % des données sont inférieures à celle-ci et 50 % des données sont supérieures.*** 

+ Nous pouvons calculer cela en triant tous les points de données et en prenant celui du milieu, qui serait l’indice 41 dans ce cas. 

+ Cela nous donne une médiane de 10 points 1 heure de sommeil. En Python, nous pouvons utiliser **np.median** pour faire les calculs pour nous.

####**7. Mesures du centre: mode**

+ Le mode est la valeur la plus fréquente dans les données. 

+ Si nous comptons le nombre d’occurrences de chaque sleep_total et trions par ordre décroissant, il y a 4 mammifères qui dorment pendant 12,5 heures, c’est donc le mode. 

+ Le mode de la variable vore, qui indique le régime alimentaire de l’animal, est herbivore. 

+ Nous pouvons également trouver le mode à l’aide de la fonction mode du module de statistiques. 

+ Le mode est souvent utilisé pour les variables catégorielles, car les variables catégorielles peuvent être désordonnées et n’ont souvent pas de représentation numérique inhérente.

####**8. Ajout d’une valeur aberrante**

+ Maintenant que nous avons beaucoup de façons de mesurer le centre, comment savons-nous lequel utiliser? Prenons un exemple. 

+ Ici, nous avons tous les insectivores dans l’ensemble de données.

+ Nous obtenons un temps de sommeil moyen de 16 points 5 heures et un temps de sommeil médian de 18 points 9 heures.

####**10. Ajout d’une valeur aberrante**
+ Maintenant, disons que nous avons découvert un nouvel insectivore mystérieux qui ne dort jamais.

####**11. Ajout d’une valeur aberrante**

+ Si nous reprenons la moyenne et la médiane, nous obtenons des résultats différents. 

+ La moyenne a baissé de plus de 3 heures, tandis que la médiane a changé de moins d’une heure. En effet, la moyenne est beaucoup plus sensible aux valeurs extrêmes que la médiane.

####**12. Quelle mesure utiliser?**
+ Comme la moyenne est plus sensible aux valeurs extrêmes, elle fonctionne mieux pour des données symétriques comme celle-ci. 

+ Notez que la moyenne, en noir, et la médiane, en rouge, sont assez proches.

####**13. Inclinaison**
+ Cependant, si les données sont biaisées, ce qui signifie qu’elles ne sont pas symétriques, comme celle-ci, la médiane est généralement préférable à utiliser. 

+ Dans cet histogramme, les données sont empilées à droite, avec une queue à gauche. 

+ Les données qui ressemblent à ceci sont appelées données asymétriques à gauche. 

+ Lorsque les données sont empilées à gauche avec une queue à droite, elles sont biaisées vers la droite.

####**14. Quelle mesure utiliser?**


+ Lorsque les données sont faussées, **la moyenne et la médiane sont différentes**. 

+ La moyenne est tirée dans la direction de l’inclinaison, elle est donc inférieure à la médiane sur les données asymétriques à gauche et supérieure à la médiane sur les données asymétriques à droite. 

+ Parce que la moyenne est tirée par les valeurs extrêmes, il est préférable d’utiliser la médiane car elle est moins affectée par les valeurs aberrantes.

####**2. Qu’est-ce que la propagation?**

+ La propagation est exactement ce à quoi cela ressemble - elle décrit à quel point les points de données sont séparés ou rapprochés. 

+ Tout comme les mesures du centre, il existe quelques mesures différentes de la propagation.

####**3. Écart**
+ La première mesure, la variance, mesure la moyenne

####**4. Écart**
+ distance entre chaque point de données et la moyenne des données.

####**5. Calcul de la variance**

+ Pour calculer la variance, nous commençons par calculer la distance entre chaque point et la moyenne, de sorte que nous obtenons un nombre pour chaque point de données. 

+ Nous mettons ensuite chaque distance au carré

####**6. Calcul de la variance**

+ puis additionnez-les tous ensemble. 

+ Enfin, nous divisons la somme des distances au carré par le nombre de points de données moins 1, ce qui nous donne la variance. 

+ Plus la variance est élevée, plus les données sont dispersées. Il est important de noter que les unités de variance sont au carré, donc dans ce cas, c’est 19 points-8 heures au carré. 

+ Nous pouvons calculer la variance en une seule étape en utilisant **np.var**, en définissant l’argument ddof sur 1. 

+ Si nous ne spécifions pas que ddof est égal à 1, une formule légèrement différente est utilisée pour calculer la variance qui ne devrait être utilisée que sur une population complète, pas sur un échantillon.

####**7. Écart type**

+ L’écart-type est une autre mesure de l’écart, calculée en prenant la racine carrée de la variance. 

+ Il peut être calculé à l’aide de **np.std**. 
+ Tout comme **np.var**, nous devons définir ddof sur 1. 

+ La bonne chose à propos de l’écart-type est que les unités sont généralement plus faciles à comprendre car elles ne sont pas au carré. Il est plus facile d’envelopper votre tête autour de 4 heures et demie que 19 points-8 heures au carré.

####**8. Écart absolu moyen**
+ L’écart absolu moyen prend la valeur absolue des distances à la moyenne, puis prend la moyenne de ces différences. 

+ Bien que cela soit similaire à l’écart-type, ce n’est pas exactement la même chose. 

+ L’écart-type carré les distances, de sorte que les distances plus longues sont pénalisées plus que les distances plus courtes, tandis que l’écart absolu moyen pénalise chaque distance de manière égale. 

+ L’un n’est pas meilleur que l’autre, mais SD est plus commun que MAD.

####**9. Quantiles**

+ Avant de discuter de la prochaine mesure de propagation, parlons rapidement des quantiles. 

+ Les quantiles, également appelés percentiles, divisent les données en un certain nombre de parties égales. 

+ Ici, nous appelons **np.quantile**, en passant dans la colonne d’intérêt, suivi du point-5. 

+ Cela nous donne 10 points 1 heure, donc 50% des mammifères de l’ensemble de données dorment moins de 10 points 1 heure par jour, et les 50% restants dorment plus de 10 points 1 heure, donc c’est exactement la même chose que la médiane. 

+ Nous pouvons également passer une liste de nombres pour obtenir plusieurs quantiles à la fois. 

+ Ici, nous divisons les données en 4 parties égales. Ceux-ci sont également appelés quartiles. Cela signifie que 25% des données se situent entre 1 point 9 et 7 points 85, 25% entre 7 points 85 et 10 points 10, et ainsi de suite.

####**10. Les boxplots utilisent des quartiles**

+ Les cases dans les diagrammes en boîte représentent des quartiles. 

+ Le bas de la boîte est le premier quartile, et le haut de la boîte est le troisième quartile. 

+ La ligne médiane est le deuxième quartile, ou médiane.

####**11. Quantiles utilisant np.linspace()**

+ Ici, nous divisons les données en cinq parties égales, mais nous pouvons également utiliser **np.linspace** comme raccourci, qui prend en compte le nombre de départ, le numéro d’arrêt et les intervalles numériques. 

+ Nous pouvons calculer les mêmes quantiles en utilisant np-dot-linspace en commençant à zéro, en s’arrêtant à un, en se divisant en 5 intervalles différents.

####**12. Intervalle interquartile (IQR)**


+ L’intervalle interquartile, ou IQR, est une autre mesure de la propagation. 

+ C’est la distance entre le 25e et le 75e percentile, qui est également la hauteur de la boîte dans un boxplot. 

+ Nous pouvons le calculer en utilisant la fonction quantile, ou en utilisant la fonction iqr de scipy-dot-stats pour obtenir 5 points-9 heures.

####**13. Valeurs aberrantes**
+ **Les valeurs aberrantes sont des points de données qui sont sensiblement différents des autres.** 

+ Mais comment savons-nous ce qu’est une différence substantielle? 

+ ***Une règle souvent utilisée est que tout point de données inférieur au premier quartile moins 1.5 fois l’IQR est une valeur aberrante, ainsi que tout point supérieur au troisième quartile plus 1.5 fois l’IQR.***

####**9. Distribution binomiale**
+ ***La distribution binomiale décrit la probabilité du nombre de succès dans une séquence d’essais indépendants.*** 

+ En d’autres termes, il peut nous indiquer la probabilité d’obtenir un certain nombre de têtes dans une séquence de pile ou face. 

+ Notez qu’il s’agit d’une distribution discrète puisque nous travaillons avec un résultat dénombrable. 

+ La distribution binomiale peut être décrite à l’aide de deux paramètres, $n$ et $p$. 

+ $n$ représente le nombre total d’essais réalisés, et $p$ est la probabilité de succès. 

+ $N$ et $P$ sont également les troisième et deuxième arguments de **Binom-Dot-RVS**.

####**8. Théorème central limite**
+ Ce phénomène est connu sous le nom de théorème central limite, qui stipule qu’une distribution d’échantillonnage se rapprochera d’une distribution normale à mesure que le nombre d’essais augmente. 

+ Dans notre exemple, la distribution d’échantillonnage s’est rapprochée de la distribution normale à mesure que nous prenions de plus en plus de moyennes d’échantillonnage. 

+ Il est important de noter que le théorème central limite ne s’applique que lorsque les échantillons sont prélevés au hasard et sont indépendants, par exemple, en choisissant au hasard des accords de vente avec remplacement.

####**12. Moyenne de distribution de l’échantillonnage**

+ Le théorème central limite est également utile lorsque vous avez une population énorme et que vous n’avez pas le temps ou les ressources nécessaires pour collecter des données sur tout le monde.

###**La distribution de Poisson**

####**1. La distribution de Poisson**
+ Dans cette vidéo, nous allons parler d’une autre distribution de probabilité appelée **distribution de Poisson**.

####**2. Processus de Poisson**
+ Avant de parler de probabilité, définissons un processus de Poisson. 

+ Un processus de Poisson est un processus où les événements semblent se produire à un certain rythme, mais complètement aléatoire. 

+ Par exemple, le nombre d’animaux adoptés d’un refuge pour animaux chaque semaine est un processus de Poisson - on sait peut-être qu’en moyenne il y a 8 adoptions par semaine, mais ce nombre peut différer aléatoirement. 

+ D’autres exemples seraient le nombre de personnes arrivant dans un restaurant chaque heure, ou le nombre de tremblements de terre par an en Californie. 

+ L’unité de temps comme les heures, les semaines ou les années n’est pas pertinente tant qu’elle est cohérente.

####**3. Poisson distribution**
+ La distribution de Poisson décrit la probabilité qu’un certain nombre d’événements se produisent sur une période de temps déterminée. 

+ Nous pouvons utiliser la distribution de Poisson pour calculer la probabilité qu’au moins 5 animaux soient adoptés en une semaine, la probabilité que 12 personnes arrivent dans un restaurant en une heure ou la probabilité de moins de 20 tremblements de terre en Californie en un an.

####**4. Lambda ($\lambda$)**
+ La distribution de Poisson est décrite par une valeur appelée lambda, qui représente le nombre moyen d’événements par période. 

+ Dans l’exemple des refuges pour animaux, il s’agirait du nombre moyen d’adoptions par semaine, soit 8. 

+ Cette valeur est aussi la valeur attendue de la distribution ! 

+ La distribution de Poisson avec lambda égal à 8 ressemble à ceci. 

+ Notez qu’il s’agit d’une distribution discrète puisque nous comptons les événements, et 7 et 8 sont le nombre d’adoptions le plus probable en une semaine.

####**5. Lambda est le pic de distribution**
+ Lambda change la forme de la distribution, de sorte qu’une distribution de Poisson avec lambda égale à 1, en bleu, est très différente d’une distribution de Poisson avec lambda égal à 8, en vert, mais quoi qu’il arrive, le pic de la distribution est toujours à sa valeur lambda.

####**6. Probabilité d’une valeur unique**

+ Étant donné que le nombre moyen d’adoptions par semaine est de 8, quelle est la probabilité de 5 adoptions par semaine? 

+ Tout comme les autres distributions de probabilités, nous pouvons importer poisson à partir de **scipy-dot-stats**. 

+ Nous utiliserons la fonction **poisson-point-pmf**, en passant 5 comme premier argument et 8 comme deuxième argument pour indiquer la moyenne de la distribution. Cela nous donne environ 9%.

####**7. Probabilité inférieure ou égale à**
+ Pour obtenir la probabilité que 5 adoptions ou moins se produisent en une semaine, utilisez la fonction **poisson-point-cdf**, en passant les mêmes nombres. Cela nous donne environ 20%.

####**8. Probabilité supérieure à**
+ Tout comme les autres fonctions de probabilité que vous avez apprises jusqu’à présent, prenez 1 moins la probabilité « inférieure ou égale à 5 » pour obtenir la probabilité de plus de 5 adoptions. 

+ Il y a 81% de chances que plus de 5 adoptions aient lieu. 

+ Si le nombre moyen d’adoptions passe à 10 par semaine, il y aura 93% de chances que plus de 5 adoptions aient lieu.

####**9. Échantillonnage à partir d’une distribution de Poisson**
+ Tout comme les autres distributions, nous pouvons prélever des échantillons des distributions de Poisson en utilisant **poisson-dot-rvs**. 

+ Ici, nous allons simuler 10 semaines différentes au refuge pour animaux. En une semaine, il y a 14 adoptions, mais seulement 6 dans une autre.

####**10. Le CLT s’applique toujours!**
+ Tout comme les autres distributions, la distribution d’échantillonnage des moyennes d’échantillonnage d’une distribution de Poisson semble normale avec un grand nombre d’échantillons.

####***1. Plus de distributions de probabilité**

####***2. Distribution exponentielle**
+ La première distribution est la distribution exponentielle, qui représente la probabilité qu’un certain temps passe entre les événements de Poisson. 

+ Nous pouvons utiliser la distribution exponentielle pour prédire, par exemple, la probabilité de plus de 1 jour entre les adoptions, la probabilité de moins de 10 minutes entre les arrivées au restaurant et la probabilité que 6 à 8 mois s’écoulent entre les tremblements de terre. 

+ Tout comme la distribution de Poisson, l’unité de temps n’a pas d’importance tant qu’elle est cohérente. 

+ La distribution exponentielle utilise la même valeur lambda, qui représente le taux, que la distribution de Poisson. 

+ Notez que lambda et rate signifient la même valeur dans ce contexte. Il est également continu, contrairement à la distribution de Poisson, puisqu’il représente le temps.

####**3. Demandes de service à la clientèle**

+ Par exemple, disons qu’un ticket de service client est créé toutes les 2 minutes. 

+ Nous pouvons reformuler cela pour qu’il soit en termes d’intervalle de temps d’une minute, de sorte que la moitié d’un ticket est créée chaque minute. 

+ Nous utiliserons 0,5 comme valeur lambda. La distribution exponentielle avec un taux de moitié ressemble à ceci.

####**4. Lambda en distribution exponentielle**
+ Le taux influe sur la forme de la distribution et sur la façon dont elle diminue.

####**5. Valeur attendue de la distribution exponentielle**
+ Rappelons que lambda est la valeur attendue de la distribution de Poisson, qui mesure la fréquence en termes de vitesse ou de nombre d’événements. 

+ Dans notre exemple de ticket de service client, cela signifie que le nombre attendu de demandes par minute est le point 5. 

+ La distribution exponentielle mesure la fréquence en termes de temps entre les événements. 

+ La valeur attendue de la distribution exponentielle peut être calculée en prenant 1 divisé par lambda. 

+ Dans notre exemple, le temps attendu entre les demandes est de 1 sur la moitié, ce qui est 2, il y a donc une moyenne de 2 minutes entre les demandes.

####**6. Combien de temps faut-il pour qu’une nouvelle demande soit créée?**

+ Comme pour d’autres distributions continues, nous pouvons utiliser **expon-dot-cdf** pour calculer les probabilités. 

+ La probabilité d’attendre moins de 1 minute pour une nouvelle demande est calculée à **l’aide d’expon-cdf**, en passant dans 1 suivi d’une échelle égale le point-5, ce qui nous donne environ 86% de chances. 

+ La probabilité d’attendre plus de 3 minutes peut être trouvée en utilisant 1 moins expon-cdf de 3, l’échelle est égale au point 5, ce qui donne une chance de 2%. 

+ Enfin, la probabilité d’attendre entre 1 et 3 minutes peut être trouvée en prenant expon-cdf de 3 et en soustrayant **expon-cdf de 1**. 

+ Il y a 13% de chances que vous attendiez entre 1 et 3 minutes.

####**7. Distribution t (de l’étudiant)**

+ La distribution suivante est la distribution t, qui est aussi parfois appelée distribution t de Student. 

+ Sa forme est similaire à la distribution normale, mais pas tout à fait la même. 

+ Si nous comparons la distribution normale, en bleu, avec la distribution $t$ avec un degré de liberté, en orange, les queues de la distribution $t$ sont plus épaisses. 

+ Cela signifie que dans une distribution $t$, les observations sont plus susceptibles de s’écarter davantage de la moyenne.

####**8. Degrés de liberté**
+ **La distribution $t$** a un paramètre appelé degrés de liberté, qui affecte l’épaisseur des queues de la distribution. 

+ Des degrés de liberté plus faibles se traduisent par des queues plus épaisses et un écart-type plus élevé. 

+ Au fur et à mesure que le nombre de degrés de liberté augmente, la distribution ressemble de plus en plus à la distribution normale.

####**9. Distribution log-normale**
+ La dernière distribution dont nous parlerons est la distribution log-normale. 

+ Les variables qui suivent une distribution log-normale ont un logarithme qui est normalement distribué. 

+ Il en résulte des distributions asymétriques, contrairement à la distribution normale. 

+ Il y a beaucoup d’exemples concrets qui suivent cette distribution, comme la durée des parties d’échecs, la tension artérielle chez les adultes et le nombre d’hospitalisations lors de l’épidémie de SRAS de 2003.

+ Nous pouvons également examiner les relations entre deux variables numériques à l’aide d’un nombre appelé coefficient de corrélation.

####**8. Ampleur = force de la relation**

###**Plan d’expériences**

#####**1. Plan d’expériences**
+ Souvent, les données sont créées à la suite d’une étude qui vise à répondre à une question spécifique. 

+ Cependant, les données doivent être analysées et interprétées différemment selon la façon dont les données ont été générées et la façon dont l’étude a été conçue.

####**2. Vocabulaire**


+ Les expériences visent généralement à répondre à une question sous la forme « Quel est l’effet du traitement sur la réponse ? » 

+ Dans ce contexte, le traitement fait référence à la variable explicative ou indépendante, et la réponse fait référence à la réponse ou à la variable dépendante. 

+ Par exemple, quel est l’effet d’une publicité sur le nombre de produits achetés? Dans ce cas, le traitement est une publicité et la réponse est le nombre de produits achetés.

####**3. Expériences contrôlées**


+ Dans une expérience contrôlée, les participants sont assignés au hasard au groupe de traitement ou au groupe témoin, où le groupe de traitement reçoit le traitement et le groupe témoin non. 

+ Un bon exemple de ceci est un test A / B. Dans notre exemple, le groupe de traitement verra une publicité, et le groupe témoin ne le verra pas. 

+ Outre cette différence, les groupes devraient être comparables afin que nous puissions déterminer si le fait de voir une publicité incite les gens à acheter plus. 

+ Si les groupes ne sont pas comparables, cela pourrait entraîner une confusion ou un biais. 

+ Si l’âge moyen des participants du groupe de traitement est de 25 ans et l’âge moyen des participants du groupe témoin est de 50 ans, l’âge pourrait être un facteur de confusion potentiel si les jeunes sont plus susceptibles d’acheter plus, ce qui rendra l’expérience biaisée en faveur du traitement.

####**4. L’étalon-or des expériences utilisera...**

+ L’étalon-or, ou l’expérience idéale, éliminera autant de biais que possible en utilisant certains outils. 

+ Le premier outil pour aider à éliminer les biais dans les expériences contrôlées est d’utiliser un essai contrôlé randomisé. 

+ Dans un essai contrôlé randomisé, les participants sont assignés au hasard au groupe de traitement ou au groupe témoin et leur affectation n’est basée sur rien d’autre que le hasard. Une affectation aléatoire comme celle-ci permet de s’assurer que les groupes sont comparables. 

+ La deuxième façon est d’utiliser un placebo, qui est quelque chose qui ressemble au traitement, mais n’a aucun effet. 

+ De cette façon, les participants ne savent pas s’ils font partie du groupe de traitement ou du groupe témoin. 

+ Cela garantit que l’effet du traitement est dû au traitement lui-même, et non à l’idée d’obtenir le traitement. Ceci est courant dans les essais cliniques qui testent l’efficacité d’un médicament. 

+ Le groupe témoin recevra toujours une pilule, mais c’est une pilule de sucre qui a des effets minimes sur la réponse.

####**5. L’étalon-or des expériences utilisera...**


+ Dans une expérience en double aveugle, la personne qui administre le traitement ou exécute l’expérience ne sait pas non plus si elle administre le traitement réel ou le placebo. 

+ Cela protège contre les biais dans la réponse ainsi que dans l’analyse des résultats. 

+ Ces différents outils se résument tous au même principe : s’il y a moins de possibilités de biais de s’infiltrer dans votre expérience, plus vous pouvez conclure de manière fiable si le traitement affecte la réponse.

####**6. Études observationnelles**


+ L’autre type d’étude dont nous discuterons est l’étude observationnelle. 

+ Dans une étude observationnelle, les participants ne sont pas répartis au hasard dans des groupes. 

+ Au lieu de cela, les participants s’attribuent eux-mêmes, généralement en fonction de caractéristiques préexistantes. 


+ Ceci est utile pour répondre à des questions qui ne sont pas propices à une expérience contrôlée. 


+ Si vous voulez étudier l’effet du tabagisme sur le cancer, vous ne pouvez pas forcer les gens à commencer à fumer. 


+ De même, si vous voulez étudier comment le comportement d’achat passé affecte si quelqu’un va acheter un produit, vous ne pouvez pas forcer les gens à avoir certains comportements d’achat passés. 

+ Parce que l’assignation n’est pas aléatoire, il n’y a aucun moyen de garantir que les groupes seront comparables dans tous les aspects, de sorte que les études observationnelles ne peuvent pas établir la causalité, seulement l’association. 

+ Les effets du traitement peuvent être confondus par des facteurs qui ont amené certaines personnes dans le groupe témoin et certaines personnes dans le groupe de traitement. 

+ Cependant, il existe des moyens de contrôler les facteurs de confusion, ce qui peut aider à renforcer la fiabilité des conclusions sur l’association.

####**7. Études longitudinales et transversales**

+ La dernière distinction importante à faire est entre les études longitudinales et transversales. 

+ Dans une étude longitudinale, les mêmes participants sont suivis sur une période de temps pour examiner l’effet du traitement sur la réponse. 

+ Dans une étude transversale, les données sont recueillies à partir d’un seul instantané dans le temps. 

+ Si vous vouliez étudier l’effet de l’âge sur la taille, une étude transversale mesurerait la taille des personnes d’âges différents et les comparerait. 

+ Cependant, les résultats seront confondus par l’année de naissance et le mode de vie, car il est possible que chaque génération grandisse. 

+ Dans une étude longitudinale, les mêmes personnes verraient leur taille enregistrée à différents moments de leur vie, de sorte que la confusion est éliminée. 

+ Il est important de noter que les études longitudinales sont plus coûteuses et prennent plus de temps à réaliser, tandis que les études transversales sont moins chères, plus rapides et plus pratiques.
