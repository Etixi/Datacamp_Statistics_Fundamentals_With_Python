###**Ajustement d’une régression linéaire**

####**1. Ajustement d’une régression linéaire**

+ Vous avez peut-être remarqué que les lignes de tendance de régression linéaire dans les nuages de points étaient des lignes droites.

+ C’est une caractéristique déterminante d’une régression linéaire.

####**2. Les lignes droites sont définies par deux choses**

+ Les lignes droites sont complètement définies par deux propriétés.

+ L’interception est la valeur $y$ lorsque $x$ est égal à **zéro**.

+ La pente est la pente de la ligne, égale à la quantité $y$ augmente si vous augmentez $x$ de un.

+ L’équation pour une ligne droite est que la valeur $y$ est l’interception plus la pente multipliée par la valeur $x$.

####**3. Estimation de l’interception**
+ Voici la ligne de tendance de l’ensemble de données d’assurance suédois. Essayons d’estimer l’interception.

####**4. Estimation de l’interception**
+ Pour trouver l’interception, regardez où la ligne de tendance coupe l’axe $y$.

####**5. Estimation de l’interception**
+ C’est moins de la moitié de la barre des cinquante, donc je suppose que c’est environ vingt.

####**6. Estimation de la pente**
+ Pour estimer la pente, nous avons besoin de deux points. Pour faciliter la devinette, j’ai choisi des points où la ligne est proche du quadrillage.

####**7. Estimation de la pente**
+ Tout d’abord, nous calculons la variation des valeurs $y$ entre les points.

+ Une valeur $y$ est d’environ quatre cents et l’autre d’environ cent cinquante, donc la différence est de deux cent cinquante.

####**8. Estimation de la pente**

+ Maintenant, nous faisons la même chose pour l’axe des $x$.

+ Un point est à cent dix, l’autre à quarante. La différence est donc de soixante-dix.

####**9. Estimation de la pente**
+ Pour estimer la pente, nous divisons un nombre par l’autre.

+ Deux cent cinquante divisés par soixante-dix font environ trois virgule cinq, c’est donc notre estimation pour la pente. Exécutons une régression linéaire pour vérifier notre supposition.

####**10. Exécution d’un modèle**

+ Pour exécuter un modèle de régression linéaire, vous importez la fonction ols à partir de statsmodels dot formula dot api.

+ ***OLS signifie moindres carrés ordinaires, qui est un type de régression, et est couramment utilisé. La fonction ols prend deux arguments***.

+ Le premier argument est une formule :
  + **la variable de réponse** est écrite à gauche du tilde, et **la variable explicative** est écrite à droite.
  + L’argument **data** prend le DataFrame contenant les variables.
  + Pour réellement ajuster le modèle, vous ajoutez **la méthode .fit()** à votre objet de modèle fraîchement créé.
  + Lorsque vous imprimez le modèle obtenu, il est utile d’utiliser **l’attribut params**, qui contient les paramètres du modèle. **Il en résultera deux coefficients**.
  + Ces coefficients sont **l’interception et la pente de la ligne droite**.
  + Il semble que nos suppositions étaient assez proches.
  + L’interception est très proche de notre estimation de vingt.
  + La pente, indiquée ici comme n_claims, est de trois points quatre, légèrement inférieure à ce que nous avions deviné.

####**11. Interprétation des coefficients du modèle**
+ Cela signifie que nous nous attendons à ce que le paiement total soit vingt plus trois virgule quatre fois le nombre de réclamations.

+ Donc, pour chaque réclamation supplémentaire, nous nous attendons à ce que le paiement total augmente de trois points quatre.

####**3. Visualisation de 1 variable numérique et 1 variable catégorielle**
+ Pour visualiser les données, les nuages de points ne sont pas idéaux car les espèces sont catégoriques.

+ Pour donner un panneau séparé à chaque espèce, j’utilise la fonction de **displot de seaborn**.

####**4. Statistiques sommaires: masse moyenne par espèce**
+ Calculons quelques statistiques sommaires.

+ Nous regroupons d’abord par espèce, puis nous calculons leurs masses moyennes.

+ Vous pouvez voir que la masse moyenne d’une brème est de six cent dix-huit grammes.

+ La masse moyenne d’une perche est de trois cent quatre-vingt-deux grammes, et ainsi de suite.

####**5. Régression linéaire**
+ Exécutons une régression linéaire en utilisant la masse comme variable de réponse et l’espèce comme variable explicative.

+ La syntaxe est la même : vous appelez **ols()**, en passant une formule avec la variable de réponse à gauche et la variable explicative à droite, et en définissant l’argument de données sur DataFrame.

+ Nous ajustons le modèle à l’aide de la méthode d’ajustement et récupérons les paramètres à l’aide de paramètres de points sur le modèle ajusté.

+ Cette fois, nous avons quatre coefficients : ***un d’interception et un pour trois des espèces de poissons.***

+ Il manque un coefficient pour la brème, mais le nombre pour l’interception semble familier.

+ L’interception est la masse moyenne de la daurade que vous venez de calculer.

+ Vous vous demandez peut-être quels sont les autres coefficients et pourquoi la perche a un coefficient négatif, puisque les masses de poissons ne peuvent pas être négatives.

####**6. Modèle avec ou sans interception**

+ Les coefficients de chaque catégorie sont calculés par rapport à l’interception.

+ Cette façon d’afficher les résultats peut être utile pour les modèles avec plusieurs variables explicatives, mais pour une régression linéaire simple, c’est juste déroutant. Heureusement, nous pouvons y remédier.

+ En modifiant légèrement la formule pour ajouter « plus zéro », nous spécifions que tous les coefficients doivent être donnés par rapport à zéro.

+ De manière équivalente, cela signifie que nous ajustons une régression linéaire sans terme d’interception.

+ Si vous soustrayez deux cent trente-cinq virgule cinquante-neuf de six cent dix-sept point quatre-vingt-trois, vous obtenez trois cent quatre-vingt-deux virgule vingt-quatre, ce qui est la masse moyenne d’un perchoir.

+ Maintenant, ces coefficients ont plus de sens. Ce ne sont que les masses moyennes de chaque espèce. C’est un résultat ennuyeux et rassurant.

+ Lorsque vous n’avez qu’une seule variable explicative catégorielle, les coefficients de régression linéaire sont simplement les moyennes de chaque catégorie.

###**Régression vers la moyenne**

####**1. Régression vers la moyenne**

+ Prenons une courte pause de **la réflexion sur la modélisation de régression à un concept connexe appelé « régression vers la moyenne »**.

+ ***La régression vers la moyenne est une propriété des données, pas un type de modèle, mais la régression linéaire peut être utilisée pour quantifier son effet.***

####**2. Le concept**
+ Vous avez déjà vu que chaque valeur de réponse dans votre jeu de données est égale à la somme d’une valeur ajustée, c’est-à-dire la prédiction par le modèle, et d’un résidu, qui correspond à la quantité de données manquées par le modèle.

+ En gros, ces deux valeurs sont les parties de la réponse que vous avez expliquées pourquoi elle a cette valeur, et les parties que vous n’avez pas pu expliquer avec votre modèle.

+ Il y a deux possibilités pour expliquer pourquoi vous avez un résidu.

+ Tout d’abord, c’est peut-être simplement parce que votre modèle n’est pas génial.

+ En particulier dans le cas d’une régression linéaire simple où vous n’avez qu’une seule variable explicative, il y a souvent place à l’amélioration.

+ Cependant, il n’est généralement pas possible ou souhaitable d’avoir un modèle parfait parce que le monde contient beaucoup de hasard, et votre modèle ne devrait pas capturer cela.

+ En particulier, les réponses extrêmes sont souvent dues au hasard ou à la chance.

+ Cela signifie que les extrêmes ne persistent pas dans le temps, car la chance finit par s’épuiser.

+ C’est le concept de régression vers la moyenne. Finalement, les cas extrêmes ressembleront davantage à des cas moyens.

###**Quantifier l’ajustement du modèle**

####**1. Quantifier l’ajustement du modèle**

+ Il est généralement essentiel de savoir si les prédictions de votre modèle sont absurdes.

+ Dans ce chapitre, nous examinerons les moyens de quantifier la qualité de votre modèle.

####**2. Modèles de brèmes et de perches**


+ Auparavant, vous utilisiez des modèles sur la masse par rapport à la longueur pour la brème et la perche.

+ En regardant simplement ces nuages de points, vous pouvez avoir l’impression qu’il existe une relation linéaire entre la masse et la longueur pour la brème, mais pas pour la perche.

+ ***Il serait utile de quantifier la force de cette relation linéaire.***

####**3. Coefficient de détermination**

+ **La première mesure dont nous discuterons est le coefficient de détermination.**

+ Ceci est parfois appelé « r-carré ».

+ Pour des raisons historiques ennuyeuses, il est écrit avec un $r$ minuscule pour la régression linéaire simple et un $R$ majuscule lorsque vous avez plus d’une variable explicative.

+ ***Elle est définie comme la proportion de la variance de la variable de réponse qui est prévisible à partir de la variable explicative.***

+ Nous aurons bientôt une explication lisible par l’homme.

+ Un score de un signifie que vous avez un ajustement parfait, et un score de zéro signifie que votre modèle n’est pas meilleur que le hasard.

+ Ce qui constitue un bon score dépend de votre jeu de données.

+ Un score de zéro point cinq sur une expérience psychologique peut être exceptionnellement élevé parce que les humains sont intrinsèquement difficiles à prédire, mais dans d’autres cas, un score de zéro point neuf peut être considéré comme un mauvais ajustement.

####**4. .summary()**

+ **La méthode .summary affiche plusieurs mesures de performance dans sa sortie.**

+ Le coefficient de détermination est écrit à la première ligne et intitulé « $R^2$».

+ Sa valeur est d’environ zéro-point-huit-huit.

####**5. Attribut .rsquared**

+ Étant donné que la sortie du résumé des points n’est pas facile à utiliser, une meilleure façon d’extraire la métrique consiste à utiliser l’attribut **rsquared**, qui contient la valeur r-carré sous forme de flottant.

####**6. C’est juste une corrélation au carré**

+ ***Pour la régression linéaire simple, l’interprétation du coefficient de détermination est simple.***

+ ***Il s’agit simplement de la corrélation entre les variables explicatives et de réponse, au carré.***

####**7. Erreur-type résiduelle (RSE)**

+ **La deuxième mesure que nous examinerons est l’erreur-type résiduelle, ou RSE.**

+ ***Rappelons que chaque résidu est la différence entre une valeur prédite et une valeur observée.***

+ *Le RSE est, très grossièrement, une mesure de la taille typique des résidus. C’est-à-dire à quel point les prédictions sont généralement fausses.*

+ Elle a la même unité que la variable de réponse.

+ Dans les modèles de poissons, l’unité de réponse est le gramme.

+ Une mesure connexe, mais moins couramment utilisée, est l’erreur quadratique moyenne, ou MSE.

+ Comme son nom l’indique, ***MSE est l’erreur-type résiduelle au carré.***

####**8. .mse_resid attribut**

+ La méthode récapitulative ne contient malheureusement pas le RSE.

+ Cependant, il peut être récupéré indirectement à partir de l’attribut **mse_resid, qui contient l’erreur quadratique moyenne des résidus**.

+ Nous pouvons calculer **le RSE en prenant la racine carrée de MSE**.

+ En tant que tel, le RSE a la même unité que la variable de réponse. Le RSE pour le modèle de brème est d’environ soixante-quatorze.

####**9. Calcul du RSE: résidus au carré**
+ Calculer le RSE vous-même, c’est un peu plus compliqué. Tout d’abord, vous prenez le carré de chaque résidu.

####**10. Calcul de la RSE : somme des résidus au carré**
+ Ensuite, vous prenez **la somme de ces résidus au carré**.

####**11. Calcul du RSE: degrés de liberté**

+ Vous calculez ensuite **les degrés de liberté des résidus.**

+ Il s’agit du nombre d’observations moins le nombre de coefficients du modèle.

####**12. Calcul du RSE: racine carrée du ratio**

+ Enfin, vous prenez **la racine carrée du rapport de ces deux nombres.**

+ De manière rassurante, la valeur est toujours de soixante-quatorze.

####**13. Interprétation de RSE**

+ Un RSE de soixante-quatorze signifie que la différence entre les masses de brème prévues et les masses de brèmes observées est généralement d’environ soixante-quatorze grammes.

####**14. Erreur quadratique moyenne (RMSE)**

+ Une autre mesure connexe est l’erreur quadratique moyenne.

+ Ceci est calculé de la même manière, sauf que vous ne soustrayez pas le nombre de coefficients dans l’avant-dernière étape.

+ Il effectue la même tâche que l’erreur-type résiduelle, à savoir quantifier l’inexactitude des prédictions du modèle, mais il est pire pour les comparaisons entre modèles.

+ Vous devez savoir que RMSE existe, mais vous devez généralement utiliser RSE à la place.

+ ***Si les résidus répondaient à l’hypothèse qu’ils sont normalement distribués avec zéro moyen, alors la ligne de tendance devrait suivre de près la ligne y égale zéro sur le graphique.*** Pour l’ensemble de données de brème, c’est vrai.

+ En revanche, le modèle de la perche ne répond pas à l’hypothèse. Les résidus sont supérieurs à zéro lorsque la valeur ajustée est petite ou grande et inférieurs à zéro au milieu.

####**5. Tracé Q-Q**


+ **Le deuxième diagramme de diagnostic est appelé diagramme Q-Q**.

+ **Il montre si les résidus suivent ou non une distribution normale.**

+ **Sur l’axe des abscisses, les points sont des quantiles de la distribution normale.**

+ **Sur l’axe $y$, vous obtenez les quantiles d’échantillon, qui sont les quantiles dérivés de votre jeu de données.**

+ Cela semble technique, mais l’interprétation de cette intrigue est simple.

+  **Si les points suivent le long de la ligne droite, ils sont normalement distribués**. Sinon, ils ne le sont pas.

+ Ici, la plupart des points de brème suivent de près la ligne.

+ Deux points à chaque extrême ne suivent pas la ligne.

+ Ceux-ci correspondent aux lignes de l’ensemble de données de brèmes avec les résidus les plus élevés.

+ L’ensemble de données de perche ne suit pas la ligne d’aussi près.

+ En particulier, vous pouvez voir sur le côté droit de la parcelle que les résidus sont plus grands que prévu. Cela signifie que le modèle est particulièrement mal adapté aux plus longues longueurs de perchoir.

####**6. Tracé à l’échelle**


+ ***Le troisième graphique montre la racine carrée des résidus normalisés par rapport aux valeurs ajustées. On l’appelle souvent un tracé à l’échelle, parce que c’est plus facile à dire.**

####**6. Qu’est-ce que la régression logistique ?**

+ La solution consiste à utiliser des modèles de régression logistique, qui sont un type de modèle linéaire généralisé, utilisé lorsque la variable de réponse est logique.

+ Alors que les modèles linéaires donnent des prédictions qui suivent une ligne droite, les modèles logistiques aboutissent à des prédictions qui suivent une courbe logistique, qui est en forme de S.

####**7. Régression logistique à l’aide de logit()**

+ Pour exécuter **une régression logistique**, vous avez besoin d’une nouvelle fonction de statsmodels.

+ À partir du même package **statsmodels dot formula dot api, importez la fonction logit**.

+ Cette fonction commence le processus d’ajustement d’un modèle de régression logistique à vos données.

+ Le nom de la fonction est la seule différence entre l’ajustement d’une régression linéaire et d’une régression logistique :

  + la formule et l’argument de données restent les mêmes et vous utilisez la méthode d’ajustement ponctuel pour ajuster le modèle.

  + Comme précédemment, vous obtenez deux coefficients, un pour l’interception et un pour la variable explicative numérique.

  + L’interprétation est un peu différente; Nous y reviendrons plus tard.

####**Mesure de la performance des modèles logistiques**

+ Dans ce dernier exercice, vous allez calculer manuellement la précision, la sensibilité et la spécificité.

+ Rappelez-vous les définitions suivantes :

$$
\text{accuracy} = \frac{TN + TP}{TN + FN + FP + TP}
$$

+ La précision est la proportion de prédictions qui sont correctes.

+ La sensibilité est la proportion d'observations vraies qui sont correctement prédites par le modèle comme étant vraies.

$$
\text{specificity} = \frac{TN}{TN + FP}
$$

+ La spécificité est la proportion d'observations fausses qui sont correctement prédites par le modèle comme étant fausses.


*churn, mdl_churn_vs_relationship, et conf_matrix sont disponibles.*

####**Instructions**

+ Calculer la précision du modèle.

+ Calculer la sensibilité du modèle.

+ Calculez la spécificité du modèle.

```python
# Extract TN, TP, FN and FP from conf_matrix
TN = conf_matrix[0,0]
TP = conf_matrix[1,1]
FN = conf_matrix[1,0]
FP = conf_matrix[0,1]

# Calculate and print the accuracy
accuracy = (TN + TP) / (TN + FN + FP + TP)
print("accuracy: ", accuracy)
```

```python
# Calculate and print the sensitivity
sensitivity = TP / (TP + FN)
print("sensitivity: ", sensitivity)
```

```python

# Calculate and print the specificity
specificity = TN / (TN + FP)
print("specificity: ", specificity)
```
