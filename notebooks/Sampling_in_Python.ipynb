###**Génération de nombres pseudo-aléatoires**

####**1. Génération de nombres pseudo-aléatoires**

+ Vous avez déjà vu comment utiliser un échantillon aléatoire pour obtenir des résultats similaires à ceux de la population.

####**2. Que signifie aléatoire?**

+ Il y a plusieurs significations de random en Français. 

+ Cette définition d’Oxford Languages est la plus intéressante pour nous. 

+ Si nous voulons choisir des points de données au hasard dans une population, nous ne devrions pas être en mesure de prédire quels points de données seraient sélectionnés à l’avance de manière systématique.

  + Langues d’Oxford


####**3. Nombres aléatoires vrais**

+ Pour générer des nombres vraiment aléatoires, nous devons généralement utiliser un processus physique comme lancer des pièces de monnaie ou lancer des dés. 

+ Le service Hotbits génère des nombres à partir de la désintégration radioactive, et RANDOM-dot-ORG génère des nombres à partir du bruit atmosphérique, qui sont des signaux radio générés par la foudre. 

+ Malheureusement, ces processus sont assez lents et coûteux pour générer des nombres aléatoires.

  + https://www.fourmilab.ch/hotbits
  + https://www.random.org


####**4. Génération de nombres pseudo-aléatoires**

+ Pour la plupart des cas d’utilisation, la génération de nombres pseudo-aléatoires est meilleure car elle est bon marché et rapide. 

+ Pseudo-aléatoire signifie que bien que chaque valeur semble aléatoire, elle est en fait calculée à partir du nombre aléatoire précédent. 

+ Puisque vous devez commencer les calculs quelque part, le premier nombre aléatoire est calculé à partir de ce que l’on appelle une valeur de départ. 

+ Le mot aléatoire est entre guillemets pour souligner que ce processus n’est pas vraiment aléatoire. 

+ Si nous partons d’une valeur de départ particulière, tous les chiffres futurs seront les mêmes.

####**5. Exemple de génération de nombres pseudo-aléatoires**

+ Par exemple, supposons que nous ayons une fonction pour générer des valeurs **pseudo-aléatoires** appelée **calc_next_random**. 

+ Pour commencer, nous choisissons un numéro de graine, dans ce cas, un. **calc_next_random** fait quelques calculs et en renvoie trois. 

+ Nous en introduisons ensuite trois dans **calc_next_random**, et il fait le même ensemble de calculs et en renvoie deux. 

+ Et si nous pouvons continuer à nourrir dans le dernier nombre, il retournera quelque chose d’apparemment aléatoire.

###**Échantillonnage aléatoire et systématique simple**

####**1. Échantillonnage aléatoire et systématique simple**

+ *Il existe plusieurs méthodes d’échantillonnage à partir d’une population. Dans cette vidéo, nous examinerons l’échantillonnage aléatoire simple et l’échantillonnage aléatoire systématique.*

####**2. Échantillonnage aléatoire simple**

+ Un échantillonnage aléatoire simple fonctionne comme un tirage au sort ou une loterie. 

+ Nous commençons avec notre population de billets de tombola ou de boules de loterie et les choisissons au hasard un à la fois.

####**3. Échantillonnage aléatoire simple des cafés**

+ Dans notre ensemble de données sur les évaluations du café, au lieu de billets de tombola ou de boules de loterie, la population se compose de variétés de café. 

+ Pour effectuer un échantillonnage aléatoire simple, nous en prenons au hasard, un à la fois. 

+ Chaque café a la même chance que n’importe quel autre d’être cueilli. 

+ Lors de l’utilisation de cette technique, nous pouvons parfois nous retrouver avec deux cafés qui étaient côte à côte dans l’ensemble de données, et parfois nous pouvons nous retrouver avec de grandes zones de l’ensemble de données qui n’ont pas été sélectionnées du tout.

####**4. Échantillonnage aléatoire simple avec des pandas**

+ Nous avons déjà vu comment faire un échantillonnage aléatoire simple avec des pandas. 

+ Nous appelons l’échantillon de points et définissons $n$ à la taille de l’échantillon. 

+ Nous pouvons également définir la graine en utilisant l’argument **random_state** pour générer des résultats reproductibles, tout comme nous l’avons fait pour la génération de nombres pseudo-aléatoires. 

+ Auparavant, en ne définissant pas **random_state** lors de l’échantillonnage, notre code générait un échantillon aléatoire différent chaque fois qu’il était exécuté.

####**5. Échantillonnage systématique**

+ Une autre méthode d’échantillonnage est connue sous **le nom d’échantillonnage systématique**. 
+ Celui-ci échantillonne la population à intervalles réguliers. 
+ Ici, en regardant de haut en bas et de gauche à droite dans chaque rangée, un café sur cinq est dégusté.

####**6. Échantillonnage systématique - définition de l’intervalle**

+ L’échantillonnage systématique avec des pandas est légèrement plus délicat qu’un simple échantillonnage aléatoire. 

+ La partie délicate consiste à déterminer la taille de l’intervalle entre chaque ligne pour une taille d’échantillon donnée. 

+ Supposons que nous voulions un échantillon de cinq cafés. 

+ La taille de la population est le nombre de lignes dans l’ensemble du jeu de données, et dans ce cas, elle est de mille trois cent trente-huit. 

+ ***L’intervalle est la taille de la population divisée par la taille de l’échantillon, mais comme nous voulons que la réponse soit un entier, nous effectuons une division entière avec deux barres obliques***. 

+ C’est comme la division standard, mais élimine toute partie fractionnaires. 

+ Un-trois-trois-huit divisé par cinq est en fait deux cent soixante-sept-point-six, et en écartant la partie fractionnaire laisse deux cent soixante-sept. 

+ Ainsi, pour obtenir un échantillon systématique de cinq cafés, nous sélectionnerons tous les deux cent soixante-septième café dans l’ensemble de données.

####**7. Échantillonnage systématique - sélection des rangées**

+ Pour sélectionner toutes les deux cent soixante-septième ligne, nous appelons point-iloc sur coffee_ratings et passons des points-virgules et l’intervalle, qui est 267 dans ce cas.

+  L’intervalle des deux points-points indique aux pandas de sélectionner toutes les deux cent soixante-septième ligne de zéro à la fin du DataFrame.

####**8. Le problème de l’échantillonnage systématique**

+ Il y a cependant un problème avec l’échantillonnage systématique. 

+ Supposons que nous nous intéressions aux statistiques sur l’attribut arrière-goût des cafés. 

+ Pour examiner cela, nous utilisons d’abord **reset_index** pour créer une colonne de valeurs d’index dans notre DataFrame que nous pouvons tracer. 

+ Le tracé de l’arrière-goût par rapport à l’indice montre un modèle. 

+ Les lignes précédentes ont généralement des scores d’arrière-goût plus élevés que les lignes ultérieures. 

+ Cela introduit un biais dans les statistiques que nous calculons. 

+ En général, il n’est sûr d’utiliser l’échantillonnage systématique que si une parcelle comme celle-ci n’a pas de modèle; c’est-à-dire que cela ressemble juste à du bruit.

####**9. Sécuriser l’échantillonnage systématique**

+ Pour s’assurer que l’échantillonnage systématique est sûr, nous pouvons randomiser l’ordre des rangées avant l’échantillonnage. 

+ **dot-sample** a un argument nommé frac qui nous permet de spécifier la proportion du jeu de données à renvoyer dans l’échantillon, plutôt que le nombre absolu de lignes spécifiées par $n$. 

+ La définition de **fra**c sur un échantillonne aléatoirement l’ensemble du jeu de données. 

+ En effet, cela mélange aléatoirement les lignes. 

+ Ensuite, les indices doivent être réinitialisés afin qu’ils repartent de zéro dans l’ordre. 

+ La spécification de **drop** est égale à **True** efface les index de ligne précédents et le chaînage à un autre appel **reset_index** crée une colonne contenant ces nouveaux index. 

+ Redessiner le tracé avec le jeu de données mélangé ne montre aucun modèle entre l’arrière-goût et l’index. 

+ C’est très bien, mais notez qu’une fois que nous avons mélangé les lignes, l’échantillonnage systématique est essentiellement la même chose qu’un simple échantillonnage aléatoire.

###**Introduction à l’amorçage**

####**1. Introduction à l’amorçage**
+ Jusqu’à présent, nous nous sommes principalement concentrés sur l’idée d’échantillonner sans remplacement.

####**2. Avec ou sans**

+ Échantillonner sans remplacement, c’est comme distribuer un paquet de cartes. 

+ Lorsque nous distribuons l’as de pique à un joueur, nous ne pouvons pas distribuer l’as de pique à un autre joueur. 

+ L’échantillonnage avec remplacement est comme lancer des dés. Si nous lançons un six, nous pouvons toujours obtenir un six sur le prochain rouleau. 

+ L’échantillonnage avec remplacement est parfois appelé rééchantillonnage. Nous utiliserons les termes de manière interchangeable.

####**3. Échantillonnage aléatoire simple sans remplacement**
+ Si nous prenons un échantillon aléatoire simple sans remplacement, chaque ligne de l’ensemble de données, ou chaque type de café, ne peut apparaître qu’une seule fois dans l’échantillon.

####**4. Échantillonnage aléatoire simple avec remplacement**
+ Si nous échantillonnons avec remplacement, cela signifie que chaque ligne de l’ensemble de données, ou chaque café, peut être échantillonné plusieurs fois.

####**5. Pourquoi échantillonner avec remplacement?**
+ Jusqu’à présent, nous avons traité l’ensemble de données coffee_ratings comme la population de tous les cafés. 

+ Bien sûr, il n’inclut pas tous les cafés du monde, nous pourrions donc traiter l’ensemble de données sur le café comme étant simplement un grand échantillon de cafés. 

+ Pour imaginer à quoi ressemble l’ensemble de la population, nous devons nous rapprocher des autres cafés qui ne figurent pas dans l’ensemble de données. 

+ Chacun des cafés de l’exemple de données aura des propriétés représentatives des cafés que nous n’avons pas. 

+ Le rééchantillonnage nous permet d’utiliser les cafés existants pour nous rapprocher de ces autres cafés théoriques.

####**6. Préparation des données sur le café**
+ Pour faire simple, concentrons-nous sur trois colonnes de l’ensemble de données sur le café. 

+ Pour faciliter la visualisation des lignes qui se sont retrouvées dans l’exemple, nous allons ajouter une colonne d’index de ligne appelée index à l’aide de la méthode reset_index.

####**7. Rééchantillonnage avec .sample()**

+ Pour échantillonner avec remplacement, nous appelons sample comme d’habitude, mais nous définissons l’argument replace sur True. 

+ Si vous définissez frac sur 1, vous obtenez un échantillon de la même taille que le jeu de données d’origine.

####**10. Bootstrapping**

+ Nous allons utiliser le rééchantillonnage pour une technique appelée bootstrapping. 

+ Dans un certain sens, l’amorçage est le contraire de l’échantillonnage à partir d’une population. 

+ Avec l’échantillonnage, nous traitons l’ensemble de données comme la population et passons à un échantillon plus petit. 

+ Avec l’amorçage, nous traitons l’ensemble de données comme un échantillon et l’utilisons pour constituer une population théorique. 

+ Un cas d’utilisation de l’amorçage consiste à essayer de comprendre la variabilité due à l’échantillonnage. 

+ Ceci est important dans les cas où nous ne sommes pas en mesure d’échantillonner la population plusieurs fois pour créer une distribution d’échantillonnage.

+ Nous pouvons utiliser scipy-dot-stats et appeler norm-dot-ppf pour obtenir le CDF inverse.
