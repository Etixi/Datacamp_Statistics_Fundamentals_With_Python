```python
# Import ols from statsmodels.formula.api
from statsmodels.formula.api import ols

# Fit a linear regression of price_twd_msq vs. n_convenience
mdl_price_vs_conv = ols(formula = "price_twd_msq ~ n_convenience", data = taiwan_real_estate).fit()

# Print the coefficients
print(mdl_price_vs_conv.params)
```

```python
# Fit a linear regression of price_twd_msq vs. house_age_years, no intercept
mdl_price_vs_age = ols("price_twd_msq ~ house_age_years + 0",
                        data=taiwan_real_estate).fit()

# Print the coefficients
print(mdl_price_vs_age.params)
```

```python
# Fit a linear regression of price_twd_msq vs. n_convenience plus house_age_years, no intercept
mdl_price_vs_both = ols("price_twd_msq ~ n_convenience  + house_age_years + 0", data=taiwan_real_estate).fit()

# Print the coefficients
print(mdl_price_vs_both.params)
```

####**8. np.select()**

+ NumPy a une fonction appelée select qui simplifie le processus d’obtention de valeurs en fonction des conditions. 

+ **np.select*** prend deux arguments : 
  + **une liste de conditions et une liste de choix**. 
  
+ Les deux listes doivent être de la même longueur. 

+ Vous pouvez le lire comme suit : 
  + « Si la condition 1 est remplie, prenez le premier élément dans les choix, 
  + si la condition 2 est remplie, prenez le deuxième élément dans les choix », et ainsi de suite. 
  
+ La sortie est un tableau tiré des éléments dans les choix, en fonction des conditions. 

+ C’est très abstrait, alors regardons comment nous l’utilisons pour les prédictions.

####**9. Choisir une interception avec np.select()**

+ La liste des conditions contient un énoncé de condition pour chaque espèce. 

+ Il renvoie un Vrai de Faux, que l’espèce soit Brème, Perche, Brochet ou Gardon. 

+ La liste des choix est la collection d’interceptions qui ont été extraites des coefficients du modèle.

###**Évaluation de la performance du modèle**

####**1. Évaluation de la performance du modèle**

+ ***Le grand avantage d’utiliser plus d’une variable explicative dans un modèle est que vous pouvez parfois obtenir un meilleur ajustement que lorsque vous utilisez une seule variable explicative.***

####**2. Modéliser les mesures de performance**

+ Dans le dernier cours, vous avez vu deux mesures pour mesurer la performance du modèle :

  + **le coefficient de détermination et l’erreur type résiduelle**. 
  
+ **Le coefficient de détermination, parfois appelé valeur $R^2$, mesure la qualité de la droite de prédiction de la régression correspond aux valeurs observées, et un nombre plus élevé est préférable**. 

+ **L’erreur-type résiduelle, parfois abrégée en RSE**, est - grosso modo - la taille typique des résidus. 

+ Voyons si ces mesures s’améliorent lorsque les deux variables explicatives sont incluses dans le modèle fish.

####**3. Obtention du coefficient de détermination**

+ Pour obtenir facilement le coefficient de détermination, vous pouvez utiliser l’attribut rsquared du modèle ajusté. 

+ Pour le modèle masse en fonction de la longueur, le coefficient de détermination est zéro-point-huit-deux, où zéro est le pire ajustement possible et un est un ajustement parfait. 

+ Pour le modèle masse versus espèce, le coefficient de détermination est pire, à zéro-point-deux-cinq. 

+ Pour le modèle de masse par rapport aux deux modèles, le coefficient de détermination est le plus élevé, à zéro point neuf-deux. 

+ En utilisant cette métrique, le modèle avec les deux variables explicatives est le meilleur, car il a le coefficient de détermination le plus élevé.

####**4. Coefficient de détermination ajusté**

+ L’ajout de variables plus explicatives augmente souvent le coefficient de détermination d’un modèle, mais il y a un problème. 

+ Inclure trop de variables explicatives dans votre modèle peut conduire à un phénomène appelé surajustement. 

+ C’est à ce moment-là que votre modèle est optimisé pour fournir le meilleur ajustement pour ce jeu de données particulier, mais ne reflète plus la population générale. 

+ Dans ce cas, le modèle serait surajusté s’il fonctionnait bien sur cet ensemble de données sur les poissons, mais mauvais sur un ensemble de données différent sur les poissons. 

+ Une variante métrique appelée coefficient de détermination ajusté comprend un petit terme de pénalité pour chaque variable explicative supplémentaire afin de compenser cet effet. 

+ C’est une meilleure mesure que le simple coefficient de détermination. 

+ Son équation est basée sur le coefficient de détermination ordinaire, le nombre d’observations et le nombre de variables explicatives, y compris les interactions. 

+ La pénalité est suffisamment grande pour que l’on s’inquiète si le coefficient de détermination est faible ou si le nombre de variables explicatives est une fraction importante du nombre d’observations. 

+ Pour obtenir cette métrique, nous utilisons l’attribut **rsquared_adj**.

####**5. Obtention du coefficient de détermination ajusté**

+ Pour voir l’effet de la pénalisation, regardons les coefficients non ajustés et ajustés côte à côte. 

+ Pour le modèle masse / longueur, la métrique R-carré est le point-822, tandis que le R-carré ajusté de ce même modèle est le point-821 - une infime différence. Étant donné que chaque modèle ne contient qu’une ou deux variables explicatives, l’effet est minime dans tous les modèles.

####**6. Obtention de l’erreur-type résiduelle**

+ Le code pour obtenir l’erreur type résiduelle est le même qu’auparavant. 

+ RSE n’est pas directement disponible en tant qu’attribut, mais le MSE l’est. 

+ Rappelons que le MSE est l’erreur quadratique moyenne. 

+ Par conséquent, vous prenez la racine carrée de l’attribut mse_resid pour obtenir le RSE. 

+ Le modèle masse en fonction de la longueur a un RSE d’un peu plus de cent cinquante. 

+ Le modèle masse versus espèce a un RSE d’environ trois cent treize. 

+ Enfin, la masse par rapport aux deux modèles a un RSE d’un peu plus de cent, ce qui signifie qu’il se trompe généralement de masse d’environ cent grammes. 

+ Étant donné que ce nombre est le plus bas des trois, selon cette mesure, la masse par rapport aux deux modèles est la meilleure. 

+ Cela signifie que toutes les mesures indiquent que le modèle avec deux variables explicatives est meilleur que les modèles avec une seule variable explicative.

###**Un modèle avec une interaction**

####**1. Un modèle avec une interaction**

+ Jouer avec différents modèles pour différents bits de votre jeu de données est pénible. 

+ Une meilleure solution consiste à spécifier un modèle unique contenant des interceptions et des pentes pour chaque catégorie. 

+ Ceci est réalisé en spécifiant les interactions entre les variables explicatives.

####**2. Qu’est-ce qu’une interaction?**

+ Pour comprendre l’idée des interactions entre les variables explicatives, considérons ce que nous savons de l’ensemble de données sur les poissons. 

+ Différentes espèces de poissons ont des rapports masse / longueur différents. 

+ En termes statistiques, nous pouvons dire que l’effet de cette longueur sur la masse attendue du poisson varie selon les espèces. 

+ Cela signifie que la longueur et l’espèce interagissent. 

+ Plus généralement, si l’effet d’une variable explicative sur la réponse attendue a des valeurs différentes en fonction des valeurs d’une autre variable explicative, alors ces deux variables explicatives interagissent.

####**3. Spécification des interactions**

+ Vous avez vu comment inclure plusieurs variables explicatives dans une formule en utilisant, par exemple, plus, la longueur plus l’espèce. 

+ Pour inclure une interaction entre ces variables, il vous suffit d’échanger le plus contre une fois. 

+ J’ai appelé cette syntaxe implicite parce que vous n’avez pas écrit quelles interactions sont nécessaires - statsmodels comprend cela lui-même. 

+ Habituellement, cette syntaxe concise est préférable, mais vous souhaiterez parfois documenter explicitement les interactions incluses dans le modèle. 

+ La syntaxe explicite consiste à ajouter chaque variable explicative séparée par plus, puis à ajouter un troisième terme avec les deux variables explicatives séparées par deux points. 

+ Le résultat est exactement le même, donc le choix d’une syntaxe dépend de vos préférences personnelles : aimez-vous la brièveté ou les détails ?

####**4. Exécution du modèle**

+ Voici la formule d’un modèle, ce qui donne huit coefficients. 

+ Comme vous l’avez vu dans les modèles avec une variable explicative catégorique, les coefficients sont difficiles à comprendre. 

+ Le coefficient d’interception est l’interception pour la première espèce, à savoir la brème. Le coefficient length_cm est la pente de la brème. 

+ Ensuite, le coefficient d’interception plus le coefficient de perche point T est l’intersection pour la perche. 

+ Et le coefficient de longueur plus le coefficient de perche length_cm:species T point est la pente pour la perche. C’est un gâchis.

####**5. Des coefficients plus faciles à comprendre**

+ Ironiquement, pour obtenir des coefficients plus faciles à comprendre, nous devons rendre la formule plus difficile à lire. 

+ Il s’agit du même modèle spécifié différemment. 

+ Sur le côté droit de la formule, vous pouvez voir la variable explicative catégorique, espèce, puis une interaction entre les deux variables explicatives, puis zéro pour supprimer l’interception globale. 

+ Maintenant, nous obtenons un coefficient d’interception pour chaque espèce et un coefficient de pente pour chaque espèce.

####**6. Chiffres familiers**

+ Vous avez déjà vu toutes ces valeurs de coefficients. 

+ Si nous examinons les coefficients du modèle sur les données de la brème de la vidéo précédente, vous pouvez voir que l’interception et la pente sont les mêmes que dans le modèle que nous venons de faire. 

+ Il en va de même pour les trois autres espèces. 

+ En fait, le modèle avec l’interaction est effectivement le même que l’ajustement de modèles distincts pour chaque catégorie, sauf que vous avez la commodité de ne pas avoir à gérer quatre ensembles de code.

###**Le paradoxe de Simpson**

####**1. Le paradoxe de Simpson**

+ *Ce chapitre a examiné la différence entre les modèles d’un ensemble de données et les modèles individuels pour chaque catégorie*. 

+ *Pour certains ensembles de données, cela peut conduire à un résultat non intuitif connu sous le nom de paradoxe de Simpson*.

####**2. Un paradoxe des plus ingénieux !**

+ Le paradoxe de Simpson est que la tendance donnée par un modèle sur l’ensemble de l’ensemble de données est très différente des tendances dans les sous-ensembles de données. 

+ C’est assez abstrait, alors essayons un exemple.

####**3. Données synthétiques de Simpson**

+ Voici un ensemble de données synthétiques - c’est-à-dire inventées 
  - conçues pour démontrer le paradoxe. Chaque ligne a une coordonnée $x$ et $y$, et le jeu de données est divisé en cinq groupes, étiquetés de A à E.

    + https://www.rdocumentation.org/packages/datasauRus/topics/simpsons_paradox


####**4. Régressions linéaires**

+ L’ajustement d’une régression linéaire de $y$ par rapport à $x$ à l’ensemble de données montre une pente positive d’un point sept cinq. 

+ Cependant, l’ajustement d’un modèle qui inclut le groupe et une interaction montre quelque chose de complètement différent. 

+ La rangée inférieure de coefficients contient la pente de chaque groupe. 

+ Chaque groupe a une pente négative, ce qui contredit apparemment le fait que l’ensemble de données a une pente positive. 

+ Visualisons l’ensemble de données pour essayer de réconcilier ces coefficients opposés.

####**5. Tracer l’ensemble de l’ensemble des données**

+ Il s’agit du nuage de points désormais standard avec une ligne de tendance de régression linéaire, dessinée par **regplot**.

+ Au fur et à mesure que $x$ augmente, $y$ augmente également, ce qui entraîne une pente positive sur l’ensemble de l’ensemble de données.

####**6. Tracé par groupe**

+ La modification du graphique pour colorier les lignes par groupe montre qu’au sein de chaque groupe, $y$ diminue à mesure que $x$ augmente.

+ Un corollaire utile est que vous devez décider d’une question à laquelle répondre avant de commencer à installer des modèles.

####**8. Exemple de score de test**

+ Il est assez facile de trouver des exemples où le modèle groupé est le meilleur.

####**9. Exemple de maladie infectieuse**

+ Il est plus difficile de trouver des exemples où le modèle de l’ensemble de données est plus utile que le modèle divisé par groupe. 

+ Par exemple, pour une maladie infectieuse, le taux d’infection a tendance à être plus élevé lorsque la densité de population est plus élevée.

####**10. Concilier la différence**

+ Malheureusement, la résolution de ces désaccords de modèle est désordonnée. 

+ Souvent, les modèles, y compris les groupes, contiennent des informations que vous manqueriez autrement. Les désaccords entre les modèles peuvent révéler que vous avez besoin de variables encore plus explicatives pour comprendre pourquoi ils sont différents. 

+ Enfin, je vais répéter le conseil correct mais ennuyeux: pour choisir le meilleur modèle, vous avez besoin d’informations contextuelles sur la signification de votre ensemble de données et la question à laquelle vous essayez de répondre.

####**11. Le paradoxe de Simpson dans les ensembles de données réels**

+ Un cas aussi clair du paradoxe de Simpson est très rare. Les différences subtiles entre les modèles sont plus courantes.

####**4. Une métrique pour le meilleur ajustement**

+ L’idée la plus simple pour cette métrique serait d’additionner tous les résidus. 

+ Cela ne fonctionne pas parce que certains résidus sont négatifs, de sorte qu’ils rendraient le total plus petit au lieu de plus grand. 

+ Au lieu de cela, nous faisons la chose la plus simple, qui est de quadriller chaque résidu afin qu’ils ne soient pas négatifs, puis de les additionner. 

+ Cette métrique est appelée la somme des carrés. 

+ La partie délicate consiste à déterminer quels coefficients d’interception et de pente donneront la plus petite somme de carrés.

####**5. Un détour par l’optimisation numérique**

+ Pour résoudre ce problème, il faut faire un détour par l’optimisation numérique, ce qui signifie trouver le point minimum d’une fonction. 

+ Considérons cette équation quadratique. 

+ La valeur $y$ est $x$-carré moins $x$ plus dix. 

+ Le graphique montre que le point minimum de la fonction se produit lorsque $x$ est un peu au-dessus de zéro et $y$ est un peu en dessous de dix, mais comment pouvons-nous le trouver exactement?

####**6. Utiliser le calcul pour résoudre l’équation**

+ Il est possible de résoudre ce problème avec le calcul. 

+ Ne vous inquiétez pas si cela n’a pas de sens, vous n’en aurez pas besoin pour les exercices. Vous trouvez le minimum en prenant la dérivée, en mettant cette dérivée à zéro, en réarrangeant pour $x$, puis en la substituant dans l’équation d’origine pour trouver $$y. 

+ Il obtient la bonne réponse. $x$ est zéro point cinq et $y$ est neuf virgule sept-cinq. 

+ Cependant, toutes les équations ne peuvent pas être résolues de cette manière analytique, et depuis que Newton et Liebniz ont inventé le calcul, les mathématiciens ont essayé de trouver des moyens de l’éviter. 

+ En fait, l’un des avantages d’être un scientifique des données est que vous pouvez simplement laisser Python trouver le minimum.

####**7. minimiser()**


+ Pour effectuer une optimisation numérique en Python, vous pouvez utiliser le package optimize de scipy. 

+ Dans ce cas, vous aurez besoin de la fonction de minimisation. La fonction que vous souhaitez réduire est définie ici. 

+ Il prend x comme entrée et renvoie y comme $x$-carré moins $x$ plus dix. 

+ Ensuite, vous appelez minimiser. Le premier argument est la fonction à appeler, sans parenthèses. 

+ Le deuxième argument est une première supposition de la réponse. 

+ Pour les fonctions plus compliquées, c’est parfois important, mais ici vous pouvez choisir n’importe quoi. 

+ Dans la sortie, vous voyez fun, la valeur $y$ estimée de la fonction, qui est parfaite. 

+ La valeur $x$ se trouve en bas. C’est proche de la bonne réponse de zéro-point-cinq, et si vous avez besoin d’une meilleure précision, il existe de nombreuses options avec lesquelles vous pouvez jouer pour améliorer la réponse. 

+ Les autres éléments de sortie sont des valeurs de diagnostic, dont nous n’avons pas besoin ici.

```python
x_actual =  taiwan_real_estate["n_convenience"]
y_actual = taiwan_real_estate["price_twd_msq"]
```

```python
# Complete the function
def calc_sum_of_squares(coeffs):
    # Unpack coeffs
    intercept, slope = coeffs
    # Calculate predicted y-values
    y_pred = intercept + slope * x_actual
    # Calculate differences between y_pred and y_actual
    y_diff = y_pred - y_actual
    # Calculate sum of squares
    sum_sq = np.sum(y_diff ** 2)
    # Return sum of squares
    return sum_sq
```

####**Matrice de confusion**

+ Lorsque la variable de réponse n'a que deux résultats, comme dans le cas de la désaffection, les mesures de succès du modèle sont "combien de cas de désaffection du client le modèle a-t-il correctement prédit" et "combien de cas de non-résiliation du client le modèle a-t-il correctement prédit". 

+ On peut les trouver en générant une matrice de confusion et en calculant des métriques sommaires sur celle-ci.

+ Rappelons les définitions suivantes :

  + La précision est la proportion de prédictions qui sont correctes.

  $$
  \text{accuracy} = \frac{TN + TP}{TN + FN + FP + TP}
  $$
 
  + La sensibilité est la proportion d'observations vraies qui sont correctement prédites par le modèle comme étant vraies.
 
  $$
  \text{sensitivity} = \frac{TP}{TP + FN}
  $$

  + La spécificité est la proportion d'observations fausses qui sont correctement prédites par le modèle comme étant fausses.
 
  $$
  \text{specificity} = \frac{TN}{TN + FP}
  $$

###**Comment fonctionne la régression logistique**

####**1. Comment fonctionne la régression logistique**

+ Voyons comment fonctionne la régression logistique. 

+ Le principe est le même que pour la régression linéaire : 
  + choisissez une métrique qui mesure la distance entre les réponses prédites et les réponses réelles, et optimisez cette métrique.

####**2. La somme des carrés ne fonctionne pas**

+ Dans le cas de la régression linéaire, la métrique à optimiser était la somme des carrés. 

+ C’est-à-dire que vous avez calculé chaque réponse prévue moins la réponse réelle correspondante, l’avez mise au carré, puis avez pris la somme. 

+ Dans le cas d’une régression logistique, la réponse réelle est toujours nulle ou un, et la réponse prédite se situe entre ces deux valeurs. 

+ Il s’avère que la métrique de la somme des carrés optimise mal sous ces restrictions, et qu’il existe une meilleure métrique.

####**3. Probabilité**


+ C’est la mesure de vraisemblance. 

+ Contrairement à la somme des carrés, où le but était de trouver la valeur minimale possible, avec probabilité vous voulez trouver la valeur maximale. 

+ Vous prenez le produit des réponses prévues et réelles,

####**4. Probabilité**
+ et additionner le produit de un moins les réponses prévues et un moins les réponses réelles.

####**5. Probabilité**

+ Ensuite, vous additionnez tous les points de données. 

+ Puisque la réponse réelle n’a que deux valeurs possibles, cette équation se simplifie de deux manières différentes. 

+ Lorsque la réponse réelle est un, l’équation de chaque observation se simplifie à la réponse prévue, y_pred. 

+ Au fur et à mesure que y_pred augmente, la métrique augmente également, et la probabilité maximale se produit lorsque y_pred est un, le même que la valeur réelle. 

+ Lorsque la réponse réelle est nulle, l’équation se simplifie à un moins la réponse prévue. 

+ À mesure que y_pred diminue, la métrique augmente et la probabilité maximale se produit lorsque y_pred est nulle. 

+ Dans les deux cas, vous obtenez un score de probabilité plus élevé lorsque la réponse prévue est proche de la réponse réelle.

####**6. Log-vraisemblance**

+ Lors du calcul de la probabilité, y_pred est souvent proche de zéro ou d’un, ce qui signifie que vous finissez par additionner beaucoup de très petits nombres, ce qui introduit une erreur numérique. 

+ Il est plus efficace de calculer le logarithme de vraisemblance. 

+ La seule différence dans cette équation est que vous prenez le logarithme des termes de réponse prédits. 

+ L’optimisation pour trouver le log-vraisemblance donne les mêmes coefficients que l’optimisation pour trouver la probabilité.

####**7. Log-vraisemblance négative**

+ Puisque nous voulons maximiser la probabilité, mais que le package d’optimisation ne peut que minimiser les fonctions, une modification à faire consiste à calculer la probabilité logarithmique négative. 

+ Autrement dit, incluez un signe moins lorsque vous calculez la somme de la contribution de vraisemblance de chaque observation.

####**Algorithme de régression logistique**

+ Entrons dans les rouages et implémentons un algorithme de régression logistique. 

+ ***La fonction logit() de statsmodels étant très complexe, nous nous contenterons d'implémenter une régression logistique simple pour un seul ensemble de données.***

+ Plutôt que d'utiliser la somme des carrés comme métrique, nous voulons utiliser la vraisemblance. 

+ Cependant, la log-vraisemblance est plus stable sur le plan informatique, nous l'utiliserons donc à la place. 

+ En fait, il y a un autre changement : 
  + puisque nous voulons maximiser la log-vraisemblance, mais que **minimize()** recherche par défaut les valeurs minimales, il est plus facile de calculer la log-vraisemblance négative.

+ La valeur de la log-vraisemblance pour chaque observation est la suivante

$$
log(y_{pred}) * y_{actual} + log(1 - y_{pred}) * (1 - y_{actual})
$$

+ La métrique à calculer est la somme négative de ces contributions de log-vraisemblance.

*Les valeurs explicatives (la colonne time_since_last_purchase de churn) sont disponibles en tant que x_actual. Les valeurs de réponse (la colonne has_churned de churn) sont disponibles en tant que y_actual. logistic est importé de scipy.stats, et **logit()** et minimize() sont également chargés.*

  + Calculer la log-vraisemblance comme étant le logarithme des valeurs $y$ prédites multiplié par les valeurs $y$ réelles, plus le logarithme de un moins les valeurs $y$ prédites multiplié par un moins les valeurs $y$ réelles.

  + Calculez la somme négative de la log-vraisemblance.

  + Retourner la somme négative de la log-vraisemblance.

+ 2)

  + Minimiser la somme des carrés métriques.

  + Utilisez calc_neg_log_likelihood comme fonction à minimiser.

```python
# Complete the function
def calc_neg_log_likelihood(coeffs):
    # Unpack coeffs
    intercept, slope = coeffs
    # Calculate predicted y-values
    y_pred = logistic.cdf(intercept + slope * x_actual)
    # Calculate log-likelihood
    log_likelihood = np.log(y_pred) * y_actual + np.log(1 - y_pred) * (1 - y_actual)
    # Calculate negative sum of log_likelihood
    neg_sum_ll = -np.sum(log_likelihood)
    # Return negative sum of log_likelihood
    return neg_sum_ll

# Test the function with intercept 10 and slope 1
print(calc_neg_log_likelihood([10, 1]))
```
